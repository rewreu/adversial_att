```python
from PIL import Image
import numpy as np
import torch
from torch import Tensor
from torchvision import models, transforms
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from CW_L2_algorithm import AttackCarliniWagnerL2

%matplotlib inline
```

```python
att=AttackCarliniWagnerL2()
```

```python
model = models.resnet152(pretrained=True)
model = model.cuda()
model.training = False
model.transform_input = False
model = model.eval()

transform = transforms.Compose([
           transforms.Resize(255),
           transforms.CenterCrop(224),
           transforms.ToTensor()
        ])
img = Image.open("panda.jpg").convert('RGB')
img_=transform(img).unsqueeze_(0).cuda()
```

```python
att_img = att.run(model=model,input=img_, target=torch.zeros(1).long().cuda())
```
