```python
from PIL import Image
import numpy as np
import torch
from torch import Tensor
from torchvision import transforms
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normalize = transforms.Normalize(
   mean=[0.485, 0.456, 0.406],
   std=[0.229, 0.224, 0.225]
)

preprocess1 = transforms.Compose([
   transforms.ToTensor()
])

preprocess2 = transforms.Compose([
   normalize
])

preprocess3 = transforms.Compose([
   transforms.ToTensor(),
   normalize
])

inv_normalize = transforms.Normalize(
    mean=[-0.485/0.229, -0.456/0.224, -0.406/0.255],
    std=[1/0.229, 1/0.224, 1/0.255]
)
def tensor2numpy4plot(img_tensor):
    return (img_tensor.transpose(0,2).transpose(0,1).cpu().detach().numpy()*255.0).astype(np.int)
```

```python
img_pil = Image.open("West-Highland-White-Terrier-750x410.jpg")
img_tensor = preprocess1(img_pil)
img_arr = np.asarray(img_pil)
```

```python
img_tensor
```

```python
## test transforms.ToTensor function, it rescaled array to 
print(img_arr[100,100:120,0]/255.0,"\n")
print(img_tensor[0,100,100:120],"\n")
print(tensor2numpy4plot(img_tensor)[100,100:120,0]-img_arr[100,100:120,0])
```

```python
img_norm = normalize(img_tensor.clone())
#img_norm = preprocess2(img_tensor)
img_inv_norm = inv_normalize(img_norm.clone())
img_norm2 = preprocess3(img_pil)
img_inv_norm2 = inv_normalize(img_norm2.clone())
```

```python
print(img_norm[0,100,100:120])
print(img_norm2[0,100,100:120])
print(img_inv_norm[0,100,100:120])
print(img_inv_norm2[0,100,100:120])
```
